```python
!pip install numpy==1.20
!pip install sdv

import sdv
from sdv.tabular import CTGAN

import pandas as pd
from google.colab import files
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import io

uploaded = files.upload() 
df2 = pd.read_csv(io.BytesIO(uploaded['IMTdatabase.csv']),sep=';' , encoding='latin-1') 
```

Shuffle database

```python
import numpy as np
from numpy.random import seed
from numpy.random import rand

seed()
df2 = df2.sample(frac = 1)

df2.head()
```

Create CTGAN using library

```python
import sdv
from sdv.tabular import CTGAN

model = CTGAN(epochs=200,batch_size=150,log_frequency=True,embedding_dim=256,generator_lr=1e-5,generator_decay=1e-6,discriminator_lr=1e-5,discriminator_decay=1e-6,discriminator_steps=10   )
model.fit(df2)
```

```python
from sdv.sampling import Condition

new_data = model.sample(num_rows=5)
```
